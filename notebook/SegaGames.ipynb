```python
test1=[
'Creative Assembly',
'Sonic Team',
'Ryu ga Gotoku Studios',
'Sports Interactive',
'Sumo Digital',
'M2',
'Visual Concepts',
'Two Point Studios',
'Sega AM2',
'Backbone Entertainment',
'Atlus',
'Gearbox Software',
'Sega,  Sonic Team'
'Amplitude Studios',
'PlatinumGames',
'Hitmaker',
'Dimps Corporation',
'Wow Entertainment',
'Sega,  Ryu ga Gotoku Studios'
]
```

```python
df2=df1[(df1['meta_score']!=0)&(df1['user_score']!=0)&(df1['esrb_rating']!=0)&(df1['developers_1']!='Sega')]
```

```python
df2[df2['developers_1']=='Two Point Studios']
```

```python
df_test1=pd.DataFrame(columns=df2.columns)
```

```python
for i, row in tqdm(df2.iterrows()):
    if row['developers_1'] in test1:
        df_test1 = df_test1.append(row,ignore_index=True)
```

```python
df_test1_groupby  = df_test1.groupby(['platform','esrb_rating']).agg(users=('platform', 'count'),rating =('user_score', 'mean'))
```

```python
df_test1_groupby
```

```python
platforms = df_test1_groupby.index.levels[0]
num_platforms = len(platforms)
num_per_row = 2
num_rows = (num_platforms + num_per_row - 1) // num_per_row

fig, axes = plt.subplots(nrows=num_rows, ncols=num_per_row, figsize=(12, 6*num_rows), sharex=True)

for i, platform in enumerate(platforms):
    row = i // num_per_row
    col = i % num_per_row
    platform_data = df_test1_groupby.loc[platform]
    ax = axes[row, col]
    platform_data.plot(kind='bar', ax=ax)
    ax.set_title(f'Платформа: {platform}')
    ax.set_xlabel('ESRB Рейтинг')
    ax.set_ylabel('Количество пользователей / Средний рейтинг')
    ax.legend(['Количество пользователей', 'Средний рейтинг'])

plt.tight_layout()
plt.show()
```

```python
platforms = df_test1_groupby.index.levels[0]
num_platforms = len(platforms)
num_per_row = 2
num_rows = (num_platforms + num_per_row - 1) // num_per_row

fig, axes = plt.subplots(nrows=num_rows, ncols=num_per_row, figsize=(12, 6*num_rows), sharex=True, sharey=True)

for i, platform in enumerate(platforms):
    row = i // num_per_row
    col = i % num_per_row
    platform_data = df_test1_groupby.loc[platform]
    ax = axes[row, col]
    ax.pie(platform_data['users'], labels=platform_data.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Платформа: {platform}')

plt.tight_layout()
plt.show()
```
